```python
import pandas as pd

promotion = pd.read_csv('./data/promotion_df.csv')
```

```python
# viewed가 마감 시점(offer received + duration*24) 이후에 찍힌 경우
promotion['deadline'] = promotion['offer received'] + promotion['duration'] * 24

view_after_deadline = promotion[
    promotion['offer viewed'].notna() &
    promotion['offer received'].notna() &
    (promotion['offer viewed'] > promotion['deadline'])
].copy()

display(
    view_after_deadline[
        [
            'person', 'offer_id', 'offer_cycle',
            'offer received', 'offer viewed', 'offer completed',
            'duration', 'deadline', 'difficulty', 'amount',
            'is_received', 'is_viewed', 'is_completed', 'is_normal_flow'
        ]
    ].sort_values(['person', 'offer_id', 'offer received'])
)

print('viewed가 duration 이후인 행 수:', len(view_after_deadline))
```

```python
view_after_completed = promotion[
    promotion['offer viewed'].notna() &
    promotion['offer completed'].notna() &
    (promotion['offer viewed'] > promotion['offer completed'])
].copy()

no_view = promotion[promotion['offer viewed'].isna()].copy()

print('viewed가 completed보다 늦은 행 수:', len(view_after_completed))
print('viewed가 없는 행 수:', len(no_view))
```
