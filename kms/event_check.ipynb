```python
portfolio = pd.read_csv('../data/portfolio.csv')
profile = pd.read_csv('../data/profile.csv') # (17000, 6)
transcript = pd.read_csv('../data/transcript.csv') # (306534, 8)
```

```python
# value컬럼 뜯어보기

# 문자열 형태의 value 컬럼 -> 딕셔너리 형태로 변환
transcript['value'] = transcript['value'].apply(ast.literal_eval)

# 딕셔너리의 키 들을 새로운 컬럼으로 전개
value_df = pd.DataFrame(transcript['value'].tolist())
transcript = pd.concat([transcript, value_df], axis=1)

# offer id를 offer_id로 컬럼명 통일
transcript['offer_id'] = transcript['offer_id'].fillna(transcript['offer id'])
transcript = transcript.drop('offer id', axis=1)
```

## 본격 event 컬럼 뜯어보기

### 가설 1. offer completed는 무조건 transacion 뒤?


```python
# person_id, time순으로 정렬
df_sorted = df.sort_values(by=['person_id', 'time']).copy()

df_sorted['prev_event'] = df_sorted.groupby('person_id')['event'].shift(1)
df_sorted['prev_time'] = df_sorted.groupby('person_id')['time'].shift(1)


completed_df = df_sorted[df_sorted['event'] == 'offer completed']
print(completed_df['prev_event'].value_counts(dropna=False))
# prev_event
# transaction        30617
# offer completed     2962
# offer completed는 transaction 바로 뒤 혹은, offer completed 바로 뒤
print(f"offer completed 바로 직전의 이벤트에 대한 전체 개수 : {completed_df['prev_event'].value_counts(dropna=False).sum()}")
print("="*80)

completed_df['time_diff'] = completed_df['time'] - completed_df['prev_time']
print(completed_df['time_diff'].value_counts(dropna=False).head())
# offer completed는 곧바로 행해진다는 것을 알 수 있다.
```

- offer completed는 transaction 뒤 또는 offer completed 뒤
- transaction 뒤라면 정상. offer completed 뒤라면 offer_id가 달라야 정상

```python
df_sorted['next_event'] = df_sorted.groupby('person_id')['event'].shift(-1)

condition = (
    (df_sorted['event'] == 'offer completed') & 
    ((df_sorted['prev_event'] == 'offer completed') | (df_sorted['next_event'] == 'offer completed'))
)

multi_completed_df = df_sorted[condition]

print(f"단일 completed를 제외한, 연속 completed 이벤트 총 개수: {len(multi_completed_df)}건\n")

multi_completed_df
```

```python
same_offer_completed = multi_completed_df[
    multi_completed_df.duplicated(subset=['person_id', 'time', 'offer_id'], keep=False)
]

print(f"같은 시간에 같은 offer completed 가 들어간 총 건수: {len(same_offer_completed)}건\n")
same_offer_completed
```

- 동일 시간에 동일한 offer_id의 offer completed가 여럿 있는것이 조금 이상하다고 생각함.

```python
offer_events_df = df_sorted[df_sorted['event'].isin(['offer received', 'offer viewed', 'offer completed'])]

event_counts = offer_events_df.pivot_table(
    index=['person_id', 'offer_id'], 
    columns='event', 
    aggfunc='size', 
    fill_value=0
).reset_index()

event_counts.head(10)
```

### 가설4) 역전 순서 찾기

```python
# 1. 오퍼 관련 이벤트만 필터링 (df_sorted는 이미 person_id와 time 순으로 정렬되어 있음)
offer_events_df = df_sorted[df_sorted['event'].isin(['offer received', 'offer viewed', 'offer completed'])]

# 2. 고객(person_id)과 오퍼(offer_id)별로 그룹화하여 '가장 처음(first)' 발생한 이벤트 추출
first_events = offer_events_df.groupby(['person_id', 'offer_id']).first().reset_index()

# 3. 가장 처음 발생한 이벤트가 'offer received'가 아닌 경우 (즉, viewed나 completed가 먼저 온 경우)
time_anomalies = first_events[first_events['event'] != 'offer received']
```

```python
# 데이터 정렬 (시간 흐름대로)
offer_events_df = df.sort_values(by=['person_id', 'offer_id', 'time']).copy()

# 각 행이 지날 때마다(시간이 흐를 때마다) 누적으로 몇 번 received/completed했는지 계산
# True/False(1/0)를 누적합(cumsum)하여 그 시점까지의 총 횟수를 구한다.
offer_events_df['cum_received'] = (offer_events_df['event'] == 'offer received').groupby(
    [offer_events_df['person_id'], offer_events_df['offer_id']]
).cumsum()

offer_events_df['cum_completed'] = (offer_events_df['event'] == 'offer completed').groupby(
    [offer_events_df['person_id'], offer_events_df['offer_id']]
).cumsum()

# 어느 순간이든 '누적 completed 수'가 '누적 received 수'를 초과한 적이 있는가?
sequence_errors = offer_events_df[offer_events_df['cum_completed'] > offer_events_df['cum_received']]

print(f"시간 순서 역전 오류 데이터 총 건수: {len(sequence_errors)}건\n")
```
