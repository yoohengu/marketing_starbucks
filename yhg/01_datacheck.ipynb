```python
df_pf = pd.read_csv('../data/portfolio.csv')
df_pro = pd.read_csv('../data/profile.csv')
df_tran = pd.read_csv('../data/transcript.csv')
```

```python
def check(df, name="Data"):
    print(f"{name} 기본 정보")
    
    print("\n[1] 데이터 크기")
    display(df.shape)
    
    print("\n[2] 컬럼 정보")
    df.info()
    
    print("\n[3] 결측치 개수")
    display(
    df.isnull().sum()
    .sort_values(ascending=False)
    .to_frame("결측치 개수")
    )
    
    print("\n[4] 중복 데이터 개수")
    display(df.duplicated().sum())
```
